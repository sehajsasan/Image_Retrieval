```python
import os
import shutil
import json

# Define the paths to the gallery and query directories
gallery_dir = '/notebooks/Image_Retrieval/human_activity_retrieval_dataset/gallery'
query_dir = '/notebooks/Image_Retrieval/human_activity_retrieval_dataset/query_images'
new_gallery = "/notebooks/Image_Retrieval/EasyPositiveHardNegative-master/gallery"
new_query = "/notebooks/Image_Retrieval/EasyPositiveHardNegative-master/query"
# Load the JSON file
json_file = '/notebooks/Image_Retrieval/human_activity_retrieval_dataset/test_image_info.json'
with open(json_file, 'r') as f:
    data = json.load(f)

# Iterate through the data and organize the images
for image_name, class_name in data.items():
    # Determine the source directory (either 'gallery' or 'query')
    if os.path.exists(os.path.join(gallery_dir, image_name)):
        source_dir = gallery_dir
    elif os.path.exists(os.path.join(query_dir, image_name)):
        source_dir = query_dir
    else:
        print(f"Image {image_name} not found in both gallery and query directories.")
        continue
    
    # Define the target directory based on the class name
    if source_dir == gallery_dir:
        target_dir = os.path.join(new_gallery, class_name)
        os.makedirs(target_dir, exist_ok=True)
        
    if source_dir == query_dir:
        target_dir = os.path.join(new_query, class_name)
        os.makedirs(target_dir, exist_ok=True)
    # target_dir = os.path.join(source_dir, class_name)
    # os.makedirs(target_dir, exist_ok=True)
    
    # Move the image to the target directory
    src_path = os.path.join(source_dir, image_name)
    dest_path = os.path.join(target_dir, image_name)
    shutil.move(src_path, dest_path)
    print(f"Moved {src_path} to {dest_path}")

print("Image organization complete.")
```

```python
###### HAR resnet18 160 by 160 ###### 
Data='HAR'
src='/notebooks/Image_Retrieval/EasyPositiveHardNegative-master/dataset'

data_dict = {}
data_dict['gallery'] = {}
data_dict['query'] = {}
for p in [f'gallery',f'query']:
    for d in glob(os.path.join(src,p,'*')):
        if 'gallery' in p:
            dict_key = 'gallery'
        else:
            dict_key = 'query'
        data_dict[dict_key][os.path.basename(d)] = glob(os.path.join(src,'*',p,'*',d,'*.jpg'))

torch.save(data_dict,'data_dict_emb_test.pth')
```
